```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer #特征转换器
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn.model_selection import train_test_split
import numpy as np

#排除时间的影响
dddd["label"] = y

dddd['year'] = dddd['year'].map(lambda x:2020-x)
dddd['citations_number'] = dddd.apply(lambda x:x['citations_number']/x['year'], axis=1)
dddd['star_counts'] = dddd.apply(lambda x:x['star_counts']/x['year'], axis=1)
dddd['subscribe_count'] = dddd.apply(lambda x:x['subscribe_count']/x['year'], axis=1)
dddd['open_issues_count'] = dddd.apply(lambda x:x['open_issues_count']/x['year'], axis=1)
dddd['forks_count'] = dddd.apply(lambda x:x['forks_count']/x['year'], axis=1)

#对每一列归一化
'''
max_min_scaler = lambda x : (x-np.min(x))/(np.max(x)-np.min(x))
dddd[['influentialCitationCount']]=dddd[['influentialCitationCount']].apply(max_min_scaler)
dddd[['citations_number']]=dddd[['citations_number']].apply(max_min_scaler)
dddd[['star_counts']]=dddd[['star_counts']].apply(max_min_scaler)
dddd[['references_number']]=dddd[['references_number']].apply(max_min_scaler)
dddd[['subscribe_count']]=dddd[['subscribe_count']].apply(max_min_scaler)
dddd[['citationVelocity']]=dddd[['citationVelocity']].apply(max_min_scaler)
dddd[['year']]=dddd[['year']].apply(max_min_scaler)
dddd[['repo_size']]=dddd[['repo_size']].apply(max_min_scaler)
dddd[['readme_size']]=dddd[['readme_size']].apply(max_min_scaler)
dddd[['open_issues_count']]=dddd[['open_issues_count']].apply(max_min_scaler)
dddd[['forks_count']]=dddd[['forks_count']].apply(max_min_scaler)
'''
X = dddd[["year","citations_number","influentialCitationCount", "star_counts","subscribe_count", "citationVelocity", "readme_size","open_issues_count","forks_count"]]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=33)  # 将数据进行分割

vec = DictVectorizer(sparse=False)
X_train = vec.fit_transform(X_train.to_dict(orient='record'))   #对训练数据的特征进行提取
X_test = vec.transform(X_test.to_dict(orient='record'))         #对测试数据的特征进行提取
#转换特征后，凡是类别型型的特征都单独独成剥离出来，独成一列特征，数值型的则不变
print(vec.feature_names_)  

#3.使用决策树对测试数据进行类别预测
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train,y_train)
y_predict = dtc.predict(X_test)

feat_importance = dtc.tree_.compute_feature_importances(normalize=False)
print("feat importance = " + str(feat_importance))
#4.获取结果报告
print('Accracy:',dtc.score(X_test,y_test))
print(classification_report(y_predict,y_test,target_names=['died','servived']))
```
